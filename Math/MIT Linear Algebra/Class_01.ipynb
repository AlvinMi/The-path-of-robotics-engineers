# 第一讲 方程组的几何解释

在第一讲中, 讲述的是线性代数的引入, 课程中的教授认为方程组是引入线性代数最好的开端。

首先看最简单的二元线性代数方程组\begin{cases}2x&-y&=0\\-x&+2y&=3\end{cases} 写成矩阵形式有$\begin{bmatrix}2&-1\\-1&2\end{bmatrix}\begin{bmatrix}x\\y\end{bmatrix}=\begin{bmatrix}0\\3\end{bmatrix}$, 通常把第一个矩阵称为系数矩阵 $A$, 将第二个矩阵成为向量 $x$, 将第三个矩阵称为向量 $b$, 这样线性方程组就可以表示为: $Ax=b$。 

再看行图像, 即直角坐标系中的图像：


```python
%matplotlib inline
get_ipython().magic('matplotlib inline')
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

x = [-2, 2, -2, 2]
y = [-4, 4, 0.5, 2.5]

fig = plt.figure()
plt.axhline(y=0, c='black')
plt.axvline(x=0, c='black')

plt.plot(x[:2], y[:2], x[:2], y[2:])
plt.draw()
```

```python
plt.close(fig)
```

下面按照**列图像**观察方程组 $x\begin{bmatrix}2\\-1\end{bmatrix}+y\begin{bmatrix}-1\\2\end{bmatrix}=\begin{bmatrix}0\\3\end{bmatrix}$ 把第一个向量称作$col_1$,第二个向量称作 $col_2$, 表示第一列向量和第二列向量, 要使得式子成立, 需要第一个向量加上两倍的第二个向量, 即
$1\begin{bmatrix}2\\-1\end{bmatrix}+2\begin{bmatrix}-1\\2\end{bmatrix}=\begin{bmatrix}0\\3\end{bmatrix}$。

列图形，即在二维平面的列向量：

```python
from functools import partial

fig = plt.figure()
plt.axhline(y=0, c='black')
plt.axvline(x=0, c='black')
ax = plt.gca()
ax.set_xlim(-2.5, 2.5)
ax.set_ylim(-3, 4)

arrow_vector = partial(plt.arrow, width=0.01, head_width=0.1, head_length=0.2, length_includes_head=True)

arrow_vector(0, 0, 2, -1, color='b')
arrow_vector(0, 0, -1, 2, color='g')
arrow_vector(2, -1, -2, 4, color='g')
arrow_vector(0, 0, 0, 3, width=0.05, color='r')

plt.draw()
```

```python
# plt.close(fig)
```

如图, 蓝向量 $col_1$与绿向量(两倍的蓝向量$col_2$), 合并成红向量 $b$。
继续观察$x\begin{bmatrix}2\\-1\end{bmatrix}+y\begin{bmatrix}-1\\2\end{bmatrix}=\begin{bmatrix}0\\3\end{bmatrix}$, $col_1,col_2$ 的某种线性组合得到了向量 $b$, 那么$col_1, col_2$ 的所有线性组合能够得到什么结果？ 它们将铺满整个平面。

## 下面进入三个未知数的方程组

$\begin{cases}2x&-y&&=0\\-x&+2y&-z&=-1\\&-3y&+4z&=4\end{cases}$ ， 写作矩阵形式 $A=\begin{bmatrix}2&-1&0\\-1&2&-1\\0&-3&4\end{bmatrix},\ b=\begin{bmatrix}0\\-1\\4\end{bmatrix}$。

在三维直角坐标系中, 每一个方程将确定一个平面, 而课程中特例的三个平面会相交于一点, 这个点就是方程的解。

同理, 将方程写成**列向量**的线性组合, 观察列图像:
$x\begin{bmatrix}2\\-1\\0\end{bmatrix}+y\begin{bmatrix}-1\\2\\-3\end{bmatrix}+z\begin{bmatrix}0\\-1\\4\end{bmatrix}=\begin{bmatrix}0\\-1\\4\end{bmatrix}$。 可以看出教授特意设计的例子中最后一个列向量恰好等于等式右边的 $b$ 向量, 所以我们需要的线性组合为 $x=0,y=0,z=1$。

假设令 $b=\begin{bmatrix}1\\1\\-3\end{bmatrix}$, 则需要的线性组合为 $x=1,y=1,z=0$。

然后我们并不总能这么轻易的求出正确的线性组合, 所以教授将在下一讲介绍消元法 —— 一种线性方程组的系统性解法。

现在我们需要考虑的是, 对于任意的 $b$, 是否都能求解出 $Ax=b$？ 用列向量线性组合的观点阐述, 列向量的线性组合能否覆盖整个三维向量空间？ 对于上面的例子答案是肯定的, 这个例子中的 $A$ 是我们喜欢的矩阵类型, 但是对于另外一些矩阵, 答案则是否定的。 那么在什么情况下, 三个向量的线性组合得不到 $b$ ？

如果三个向量在同一平面上, 问题就出现了；那么他们的线性组合也一定都在这个平面上。例如: $col_3=col_1+col_2$, 那么不管怎么组合, 这三个向量的结果都会在这个平面内, 因此当 $b$ 在平面内, 方程有解, 而当 $b$ 不在平面内, 这三个列向量就无法构造出 $b$。 当学了后面的课程之后就会知道, 像这类的情况称为**奇异**、**矩阵不可逆**。

## 推广到九维空间
每个方程有九个未知数, 共九个方程, 此时已经无法从坐标系图像中描述问题了, 但是我们依然可以从求九维列向量线性组合的角度解决问题,仍然是上面的问题, 是否能得到 $b$？当然这仍取决于九个向量, 如果取一些并不相互独立的向量, 则答案是否定的, 比如取了九列但其实相当于八列, 有一列毫无贡献(这一列是前面列的某种线性组合), 则会有一部分 $b$ 无法求到。

 $Ax=b$, 是一种乘法运算, 举个例子, 取 $A=\begin{bmatrix}2&5\\1&3\end{bmatrix},\ x=\begin{bmatrix}1\\2\end{bmatrix}$，看看如何计算矩阵乘以向量。
 
 * 我们仍然使用列向量线性组合的方式, 一次计算一列。$\begin{bmatrix}2&5\\1&3\end{bmatrix}\begin{bmatrix}1\\2\end{bmatrix}=1\begin{bmatrix}2\\1\end{bmatrix}+2\begin{bmatrix}5\\3\end{bmatrix}=\begin{bmatrix}12\\7\end{bmatrix}$ 
 * 另一种方法, 使用向量内积, 矩阵第一行向量**点乘** $x$ 向量 $\begin{bmatrix}2&5\end{bmatrix}\cdot\begin{bmatrix}1&2\end{bmatrix}^T=12,\ \begin{bmatrix}1&3\end{bmatrix}\cdot\begin{bmatrix}1&2\end{bmatrix}^T=7$。
 
 当然教授建议使用第一种方法, 将 $Ax$ 看做 $A$ 列向量的线性组合。
